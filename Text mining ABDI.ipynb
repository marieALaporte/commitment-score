# Keywords preparation

```python
###Read the keyword file and create lists

##ENGLISH
### prepare the lists
Pillar1GT = []
Pillar1ST = []
Pillar2GT = []
Pillar2SP = []
Pillar3GT = []
Pillar3SP = []


### read the keywords files and create lists ---- REMOVE before the ducplicated lines
with open('keywordsABDi_July2019.txt', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile, delimiter='\t')
    ##fyi header: Pillar	Scale	Indicator	Keyword group for text mining paper (created on 9 July 2019)	Keyword_EN	Keyword_EN updated 26 July 2019	Group_ES	Keyword_ES	Group_CN	Keyword_CN
    for row in reader:
        keyword = row[5].strip()
        if "\xa0" in keyword:
            #print(keyword)
            keyword=keyword.replace("\xa0"," ") ## weird encoding issue...
        if row[0] == "Pillar 1":
            if(row[1]=="General"):
                Pillar1GT.append(keyword)
            elif (row[1]=="Specific"):
                Pillar1ST.append(keyword)
        elif row[0] == "Pillar 2":
            if(row[1]=="General"):
                Pillar2GT.append(keyword)
            elif (row[1]=="Specific"):
                Pillar2SP.append(keyword)
        elif row[0] == "Pillar 3":
            if(row[1]=="General"):
                Pillar3GT.append(keyword)
            elif (row[1]=="Specific"):
                Pillar3SP.append(keyword)
```

```python
### Create groups

group = {}

with open('keywordsABDi_July2019.txt', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile, delimiter='\t')
    
    for row in reader:
        Group_EN = row[3]
        Keyword_EN = row[5].strip()
        
        if "\xa0" in Keyword_EN:
            #print(Keyword_EN)
            Keyword_EN=Keyword_EN.replace("\xa0"," ")
        #if("variet*" in Keyword_EN):
            #Keyword_EN=Keyword_EN.replace("variet*","varietal")
        #if("function*" in Keyword_EN):
            #Keyword_EN=Keyword_EN.replace("function*","functional")
        
        group[Keyword_EN] = Group_EN 
```

```python
import pandas as pd
import pandas.io.formats.excel
```

```python
##read the output files
occ1 = pd.read_csv('occurencesPilar1.csv', sep=";").T
occ2 = pd.read_csv('occurencesPilar2.csv', sep=";").T
occ3 = pd.read_csv('occurencesPilar3.csv', sep=";").T
sen1 = pd.read_csv('sentencesPilar1.csv', sep=";", index_col=0)
sen2 = pd.read_csv('sentencesPilar2.csv', sep=";", index_col=0)
sen3 = pd.read_csv('sentencesPilar3.csv', sep=";", index_col=0)
### build the excel file
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('./results/pandas_multiple.xlsx', engine='xlsxwriter',  options={'strings_to_numbers': True})
workbook = writer.book
##
pd.io.formats.excel.header_style = None

# Write each dataframe to a different worksheet.
occ1.to_excel(writer, sheet_name='OccurencesPillar1',header=False)
occ2.to_excel(writer, sheet_name='OccurencesPillar2',header=False)
occ3.to_excel(writer, sheet_name='OccurencesPillar3',header=False)
sen1.to_excel(writer, sheet_name='SentencesPillar1')
sen2.to_excel(writer, sheet_name='SentencesPillar2')
sen3.to_excel(writer, sheet_name='SentencesPillar3')

###formatting updated from http://pbpython.com/improve-pandas-excel-output.html
number_rows = len(occ1.index)
# Define our range for the color formatting
color_range = "D2:ALF{}".format(number_rows+1)
# Add a format. Light red fill with dark red text.
format1 = workbook.add_format({'bg_color': '#FFC7CE',
                               'font_color': '#9C0006'})
worksheet = writer.sheets['OccurencesPillar1']
# Highlight the values > 0 in Red
worksheet.conditional_format(color_range, {'type':     'cell',
                                            'criteria': '>',
                                            'value':    0,
                                            'format': format1})
number_rows = len(occ2.index)
# Define our range for the color formatting
color_range = "D2:ALF{}".format(number_rows+1)
# Add a format. Light red fill with dark red text.
format1 = workbook.add_format({'bg_color': '#FFC7CE',
                               'font_color': '#9C0006'})
worksheet = writer.sheets['OccurencesPillar2']
# Highlight the values > 0 in Red
worksheet.conditional_format(color_range, {'type':     'cell',
                                            'criteria': '>',
                                            'value':    0,
                                            'format': format1})
number_rows = len(occ3.index)
# Define our range for the color formatting
color_range = "D2:ALF{}".format(number_rows+1)
# Add a format. Light red fill with dark red text.
format1 = workbook.add_format({'bg_color': '#FFC7CE',
                               'font_color': '#9C0006'})
worksheet = writer.sheets['OccurencesPillar3']
# Highlight the values > 0 in Red
worksheet.conditional_format(color_range, {'type':     'cell',
                                            'criteria': '>',
                                            'value':    0,
                                            'format': format1})


# Add a format. Light red fill with dark red text.
format1 = workbook.add_format({'bg_color': '#FFC7CE',
                               'font_color': '#9C0006'})
worksheet = writer.sheets['SentencesPillar1']
# Highlight the values > 0 in Red
worksheet.conditional_format("A2:A1000", {'type': 'text',
                                            'criteria': 'containsText',
                                            'value': 'target', 
                                            'format': format1})
# Add a format. Light red fill with dark red text.
format1 = workbook.add_format({'bg_color': '#FFC7CE',
                               'font_color': '#9C0006'})
worksheet = writer.sheets['SentencesPillar2']
# Highlight the values > 0 in Red
worksheet.conditional_format("A2:A1000", {'type': 'text',
                                            'criteria': 'containsText',
                                            'value': 'target', 
                                            'format': format1})
worksheet = writer.sheets['SentencesPillar3']
# Highlight the values > 0 in Red
worksheet.conditional_format("A2:A1000", {'type': 'text',
                                            'criteria': 'containsText',
                                            'value': 'target', 
                                            'format': format1})



# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
